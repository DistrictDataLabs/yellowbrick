```python
from yellowbrick.datasets import load_game, load_spam
from yellowbrick.classifier import ClassificationReport, ConfusionMatrix, ROCAUC, ClassPredictionError, ClassPredictionError, DiscriminationThreshold
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.metrics import accuracy_score
import matplotlib
%matplotlib inline
```

```python
classifiers = [ClassPredictionError, ClassPredictionError, ClassPredictionError, ClassPredictionError, ClassPredictionError, DiscriminationThreshold]

X,y = load_game()
X = OrdinalEncoder().fit_transform(X)
label_enc = LabelEncoder()
y = label_enc.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Normal YB Call on unfit data
reg = RandomForestClassifier() #unfitted Classifier
viz = ClassPredictionError(reg)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

# Fit a classifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
print("\n","Run on fitted classifier where is_fitted is [True, False, 'auto']","\n")

viz = ClassPredictionError(clf, is_fitted=True)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = ClassPredictionError(clf, is_fitted=False)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = ClassPredictionError(clf, is_fitted='auto')
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("\n", "Run on un_fitted classifier where is_fitted is [True, False, 'auto']","\n")

viz = ClassPredictionError(reg, is_fitted=True)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = ClassPredictionError(reg, is_fitted=False)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = ClassPredictionError(reg, is_fitted='auto')
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("\n", "Run on fitted classifier where encoder is a dict \(no classes\)","\n")
viz = ClassPredictionError(clf, encoder={0:'draw', 1:'loss', 2:'win'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on an unfitted scikit-learn classifier where `encoder` is a LabelEncoder (no `classes`)")
from sklearn.preprocessing import LabelEncoder
viz = ClassPredictionError(clf, encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a fitted scikit-learn classifier where `classes` are passed to the visualizer as a list (no `encoder`)")
viz = ClassPredictionError(clf, classes=['draw','loss','win'])
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a fitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a dictionary")
viz = ClassPredictionError(clf, classes=['draw','loss','win'], encoder={0:'draw', 1:'loss', 2:'win'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a fitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a LabelEncoder")
viz = ClassPredictionError(clf, classes=['draw','loss','win'], encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("\n", "Run on unfitted classifier where encoder is a dict \(no classes\)","\n")
viz = ClassPredictionError(reg, encoder={0:'draw', 1:'loss', 2:'win'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on an unfitted scikit-learn classifier where `encoder` is a LabelEncoder (no `classes`)")
from sklearn.preprocessing import LabelEncoder
viz = ClassPredictionError(reg, encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a unfitted scikit-learn classifier where `classes` are passed to the visualizer as a list (no `encoder`)")
viz = ClassPredictionError(reg, classes=['draw','loss','win'])
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a unfitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a dictionary")
viz = ClassPredictionError(reg, classes=['draw','loss','win'], encoder={0:'draw', 1:'loss', 2:'win'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a unfitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a LabelEncoder")
viz = ClassPredictionError(reg, classes=['draw','loss','win'], encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```
