```python
from yellowbrick.datasets import load_game, load_spam
from yellowbrick.classifier import PRCurve
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from yellowbrick.datasets import load_spam


import matplotlib
%matplotlib inline
```

```python
X, y = load_spam()
label_enc = LabelEncoder()
y = label_enc.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y)


# Normal YB Call on unfit data
reg = RandomForestClassifier() #unfitted Classifier
viz = PRCurve(reg)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```

```python
# Fit a classifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
print("\n","Run on fitted classifier where is_fitted is [True, False, 'auto']","\n")

viz = PRCurve(clf, is_fitted=True)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = PRCurve(clf, is_fitted=False)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = PRCurve(clf, is_fitted='auto')
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("\n", "Run on un_fitted classifier where is_fitted is [True, False, 'auto']","\n")

viz = PRCurve(reg, is_fitted=True)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = PRCurve(reg, is_fitted=False)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

viz = PRCurve(reg, is_fitted='auto')
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("\n", "Run on fitted classifier where encoder is a dict \(no classes\)","\n")
viz = PRCurve(clf, encoder={0:'not spam', 1:'spam'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on an unfitted scikit-learn classifier where `encoder` is a LabelEncoder (no `classes`)")
from sklearn.preprocessing import LabelEncoder
viz = PRCurve(clf, encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a fitted scikit-learn classifier where `classes` are passed to the visualizer as a list (no `encoder`)")
viz = PRCurve(clf, classes=['not spam','spam'])
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a fitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a dictionary")
viz = PRCurve(clf, classes=['no spam','spam'], encoder={0:'not spam', 1:'spam'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a fitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a LabelEncoder")
viz = PRCurve(clf, classes=['not spam','spam'], encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("\n", "Run on unfitted classifier where encoder is a dict \(no classes\)","\n")
viz = PRCurve(reg, encoder={0:'not spam', 1:'spam'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on an unfitted scikit-learn classifier where `encoder` is a LabelEncoder (no `classes`)")
from sklearn.preprocessing import LabelEncoder
viz = PRCurve(reg, encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a unfitted scikit-learn classifier where `classes` are passed to the visualizer as a list (no `encoder`)")
viz = PRCurve(reg, classes=['not spam','spam'])
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a unfitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a dictionary")
viz = PRCurve(reg, classes=['not spam','spam'], encoder={0:'not spam', 1:'spam'})
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()

print("# Call fit, score, poof on a unfitted scikit-learn classifier where `classes` are passed to the visualizer as a list and `encoder` is a LabelEncoder")
viz = PRCurve(reg, classes=['not spam','spam'], encoder=label_enc)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```
