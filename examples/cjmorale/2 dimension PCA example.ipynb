# PCA plot of Iris Data Set

```python
from yellowbrick.features.pca import PCA2D
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
params = {'scale': True, 'center': True, 'col': y}
visualizer = PCA2D(**params)
visualizer.fit(X)
visualizer.transform(X)
visualizer.poof()
```

```python
params = {'scale': False, 'center': False, 'col': y}
visualizer = PCA2D(**params)
visualizer.fit(X)
visualizer.transform(X)
visualizer.poof()
```

```python
visualizer = PCA2D()
visualizer.fit(X)
visualizer.transform(X)
visualizer.poof()
```
